```python
df=df_train_sec_reports
```

```python
df=df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0', 'cik', 'entityName'])
```

```python
df
```

```python

df=df.fillna(0)
```

```python
# The webscrapped data from the SEC reports yelded a file with more than 30GB. But full with nulls. 
# After first step of dropping columns with a defined threshold, now it is time to parse all the data so instead of a giant
# string, we have a list of dictionaries
```

```python
def parse_json_strings(df):
    parsed_data = df.applymap(lambda x: eval(x) if isinstance(x, str) else x)
    return parsed_data

```

```python
parsed_df=parse_json_strings(df)
```

```python
parsed_df.to_csv('parsed_data.csv')
```

```python
test4=parsed_df['facts.dei.EntityCommonStockSharesOutstanding.units.shares'][0]
```

```python
test4
```

```python
def get_value(l):
    if l == 0:
        return None
    d = l[len(l)-1]
    value=d['val']
    return value
    

```

```python
for col in parsed_df.columns:
    
    parsed_df[col]=parsed_df[col].apply(get_value)
    
    
```

```python
parsed_df
```

```python
threshold = 0.2  
missing_perc = parsed_df.isna().mean() 
cols_to_keep = missing_perc[missing_perc < threshold].index  
finance_data_cleaned = parsed_df[cols_to_keep] 
```
